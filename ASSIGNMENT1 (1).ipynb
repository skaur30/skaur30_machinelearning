```python
# Machine Learning Assignment 1 - Simulated Dataset using Python

# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Step 2: Generate the Simulated Dataset
n_classes = 4
# Generating 1000 samples with 4 cluster centers
data, labels = make_blobs(n_samples=1000, centers=n_classes, random_state=100)

# Step 3: Visualize the Dataset
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=30)
plt.title("Simulated Dataset with 4 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar()
plt.show()

# Step 4: Split the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Step 5: Train a Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Step 6: Make Predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate the Model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Conclusion:
# The Logistic Regression model achieved an accuracy of approximately X%.
# The classification report provides precision, recall, and F1-score for each class.
# This code can be extended by using different models, hyperparameter tuning, and additional analysis.
```
